```python
#Датасет содержит данные о среднем ежедневном времени, которое дети проводят перед экраном.

import modul_2 as modul

file_path = 'screen_time.csv'
target_column = 'Average Screen Time (hours)'
columns_drop = []
categorical_columns = {
    'Gender':{'Mal':0, 'Femal':1, 'Other/Prefer not to say':3},
    'Day Type':{'Weekend':0, 'Weekday':1},
    'Screen Time Type':{'Total':0, 'Educational':1, 'Recreational':2}}

model = modul.Time (
    file_path = file_path,
    target_column = target_column,
    columns_drop = columns_drop,
    categorical_columns = categorical_columns
)

model.load_data(file_path)
model.preprocess_data()
model.split_data (test_size=0.3, random_state=42)
model.train_model()
metrics = model.evaluate_model()
model.visualize_resul()
```
